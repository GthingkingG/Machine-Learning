# 2차원 배열 인덱싱과 슬라이싱
+)row -> 가로줄 / col -> 세로줄

```python
import numpy as np

gdp_array = np.array([
    [12257, 11561, 13165, 14673, 16496, 19403],  # 대한민국
    [39169, 34406, 32821, 35387, 38299, 37813],  # 일본
    [959, 1053, 1149, 1289, 1509, 1753],         # 중국
    [36335, 37133, 38023, 39496, 41713, 44115]   # 미국
])

print(gdp_array[1]) #일본 gdp
print(gdp_array[1, 3], gdp_array[1][3])
print(gdp_array[1:3, 2:5]) #일본, 중국의 2~4번째 gdp
print(gdp_array[:, 2:4]) #모든 나라의 2~3번째 gdp
print(gdp_array[1:3, :]) #일본, 중국의 모든 gdp
```

# 불린 인덱싱
-> Numpy Array에서 값에 접근할 때, 위치 기반이 아닌 특정 조건에 따라 인덱싱

```python
import numpy as np


gdp_array = np.array([6610, 7637, 8127, 8885, 10385, 12565, 13403, 12398, 8282, 10672])

print(gdp_array > 10000) #조건식에 따라서, 출력
print(gdp_array[gdp_array>10000]) #조건이 참인 원소만 저장
print(gdp_array) #계산과 저장은 다르다. 따로 Numpy Array에 저장하지 않으면, 원본은 변하지 않는다.

mask = gdp_array > 10000
print(mask)
print(type(mask))
print(gdp_array[mask])

print(gdp_array[(gdp_array < 8000) | (gdp_array > 10000)])
print(gdp_array[(gdp_array > 8000) & (gdp_array < 8500)])
```

+)Numpy Array에 조건식을 추가하면 그 조건에 해당하는 값이 출력되고, Numpy Array에 조건을 인덱싱하면 참인 원소만 Numpy Array에 저장된다.

# Numpy 기본연산
-> Numpy Array의 연산을 효율적으로 해주는 함수들이 존재한다.

```python
import numpy as np


gdp_korea_array = np.array([12257, 11561, 13165, 14673, 16496, 19403])
gdp_us_array = np.array([36335, 37133, 38023, 39496, 41713, 44115])

print(gdp_korea_array / 100)

print(gdp_korea_array.mean())  #평균값
print(gdp_korea_array.sum())
print(gdp_korea_array.min())
print(gdp_korea_array.max())

print(gdp_korea_array + gdp_us_array)
```

+)마찬가지로 Numpy Array에 직접 저장하지 않고 계산만하면, 원본은 변하지 않는다.
